```python
import numpy as np
import matplotlib.pyplot as plt

def generate_gaussian_samples(mean, covariance, num_samples):
    samples = np.random.multivariate_normal(mean, covariance, num_samples)
    return samples

def project_data(X, d2):
    # Estimate Covariance Matrix
    covariance_matrix = np.cov(X, rowvar=False)

    # Perform eigen-decomposition
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

    # Sort eigenvalues and corresponding eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Select the first d2 columns of the eigenvector matrix
    transformation_matrix = eigenvectors[:, :d2]

    # Project data to lower-dimensional subspace
    Y = np.dot(transformation_matrix.T, X.T).T

    return Y, eigenvalues, transformation_matrix

# Parameters for part (a)
mean_a = np.array([0, 0, 0])
covariance_a = np.array([[2, 0, 0], [0, 1.5, 0], [0, 0, 2.5]])
num_samples_a = 1000

# Generate samples
samples_a = generate_gaussian_samples(mean_a, covariance_a, num_samples_a)

# Project data to 2-dimensional space
d2 = 2
projected_data_a, eigenvalues_a, transformation_matrix_a = project_data(samples_a, d2)

# Estimate and report the covariance matrix of the projected data Y
covariance_matrix_projected_a = np.cov(projected_data_a, rowvar=False)
print("Covariance Matrix of Projected Data (Part a):\n", covariance_matrix_projected_a)

# Visualize 2-dimensional scatter plot
plt.scatter(projected_data_a[:, 0], projected_data_a[:, 1])
plt.title('2-Dimensional Scatter Plot of Projected Data (Part a)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
